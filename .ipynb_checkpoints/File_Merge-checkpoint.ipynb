```python
import pandas as pd
import os
import numpy as np



file = 'Data/51_Peg_rv.csv'
array = []
with open(file,'r') as dataFile:
        for line in dataFile:
            array.append(line.strip().split())
            
peg51 = pd.DataFrame(array)

peg51
```

```python
outarray = [['time','mnvel','errvel','tel']]
    
for i in range(len(peg51)):
    if peg51[2].to_numpy(dtype=str)[i] == 'j':
        thing = [peg51[4].to_numpy(dtype=float)[i],
                 peg51[5].to_numpy(dtype=float)[i],
                 peg51[6].to_numpy(dtype=float)[i],
                 'hires_j']
    else:
        thing = [peg51[4].to_numpy(dtype=float)[i],
                 peg51[5].to_numpy(dtype=float)[i],
                 peg51[6].to_numpy(dtype=float)[i],
                 peg51[2].to_numpy(dtype=str)[i]]
    outarray.append(thing)
    


out = pd.DataFrame(outarray)
out = out.drop(0)
# out = out.drop(11) #clear outliers
# out = out.drop(12)
# out = out.drop(13)

out = out.rename(columns={0:'time',1:'mnvel',2:'errvel',3:'tel'})
#out['mnvel'] = (out['mnvel'] - np.median(out['mnvel'].to_numpy()))*1000
#out['errvel'] = out['errvel']*1000
out = out.sort_values(by=['time'])
out
```

```python
#now let's bin this

binsize = 0.5

#(t, y, e) = np.loadtxt(infile,unpack=True,skiprows=0)
t = out['time'].to_numpy(dtype=float)
y = out['mnvel'].to_numpy(dtype=float)
e = out['errvel'].to_numpy(dtype=float)
tel = out['tel'].to_numpy()


w = 1./e**2.

#Create new binned quantities

tbin = []
ybin = []
ebin = []
telbin = []

#Bin the data

tsum = t[0]
ysum = w[0]*y[0]
wsum = w[0]
nbin = 1
telsum = tel[0]

for i in range(len(t)):
    print(t[i])

    if t[i]-t[i-1] <= binsize and tel[i] == tel[i-1]:
        nbin = nbin + 1
        tsum = tsum + t[i]
        ysum = ysum + (w[i]*y[i])
        wsum = wsum + w[i]

    else:
        tbin.append(tsum/nbin)
        ybin.append(ysum/wsum)
        telbin.append(telsum)

        if nbin == 1:
            ebin.append(e[i-1])
        else:
            #ebin.append(np.std(y[i-nbin:i]))
            ebin.append(1./np.sqrt(wsum))

        tsum = t[i]
        ysum = w[i]*y[i]
        wsum = w[i]
        nbin = 1
        telsum = tel[i]

#Catch last bin
tbin.append(tsum/nbin)
ybin.append(ysum/wsum)
telbin.append(telsum)


if nbin == 1:
    ebin.append(e[i-1])
else:
    #ebin.append(np.std(y[i-nbin:i]))
    ebin.append(1./np.sqrt(wsum))


#Write output
```

```python
#put the finally binned RVs in the data file
final = pd.DataFrame()
final['time'] = tbin
final['mnvel'] = ybin
final['errvel'] = ebin
final['tel'] = telbin
final = final.sort_values(by=['time'])
final
```

```python
final.to_csv('Files/TOI2013_rv_all.csv',sep=' ')
#also save one with no HPF
nohpf = final['tel'] != 'hpf'
final[nohpf].to_csv('Files/TOI2013_rv_nohpf.csv',sep=' ')
#also save one with carmenes only
carmonly = final['tel'] == 'carmenes'
final[carmonly].to_csv('Files/TOI2013_rv_carmenes_only.csv',sep=' ')
```
