```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import radvel
```

# RV Drive: Great for Simulating a Keplerian

```python
k1 = 7.
per1 = 21.
t1 = 2458626.69
ecc1 = 0
w1 = np.pi/2 + 0.3
tp1 = radvel.orbit.timetrans_to_timeperi(t1,per1,ecc1,w1) #need time of periastron!


orbel = [per1,tp1,ecc1,w1,k1]

t = np.linspace(0, 200, 1000)

model1 = radvel.kepler.rv_drive(t, orbel)
```

```python
plt.figure(figsize=(14,8))

plt.plot(t, model1)

plt.ylim(-10, 10)
plt.xlabel('Time', fontsize=20)
plt.ylabel('RV Amp (m/s)',fontsize=20)

plt.show()
```
